```python
# INTRODUCTION TO GIT - 

# Git is a version control system.
# It tracks changes in your code or files over time so you can:
# Save snapshots of your work
# Go back to older versions if something breaks
# Work with others on the same project without messing each other up
# Keep a history of who changed what and when
```

```python
# Why it is usefull - 

# | Without Git                      | With Git                              |
# | -------------------------------- | ------------------------------------- |
# | Hard to manage versions manually | Every version is stored automatically |
# | Risk losing work                 | Safe to undo mistakes                 |
# | Collaboration chaos              | Easy teamwork, merging branches       |
# | Sending files by zip             | Shared repository                     |

# Think of it like - 
# ðŸ“¸ A timeline of snapshots of your project, where you can jump backward or forward anytime.
```
