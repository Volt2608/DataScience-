```python
# 7 . What are constraints? Explain different types of constraints.

# Answer :
# Constraints are rules enforced on table columns to ensure data accuracy, consistency, and integrity (e.g., PRIMARY KEY, FOREIGN KEY, UNIQUE, NOT NULL, CHECK).

# PRIMARY KEY : It is used as a unique identifier, you usually want a primary key in a table you make , each
# value within a column that is set a primary key must be unique and cannot be NULL. There can only be one primary key per table.

# UNIQUE: Ensures all values are different—allows NULLs (unlike primary key).

# NOT NULL: Forces a column to always have a value.

# CHECK: Restricts values using a condition (e.g., age >= 18).

# DEFAULT: Automatically assigns a value if none is provided.

# FOREIGN KEY: is a column that links one table to another table’s primary key, enforcing referential integrity—you can’t insert invalid or orphan data.

```
