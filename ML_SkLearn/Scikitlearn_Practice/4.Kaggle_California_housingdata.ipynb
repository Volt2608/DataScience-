```python
# Downloaded the California Housing Data as CSV named - 4CH_Data.csv 
# For this Particular Data we would need to use Stratified Shuffle Split Method from scikitlearn.


# As practiced, first task is to randomly split the data into two parts, test and train in the 80-20 ratio.
# and for that we mainly use two methods from scikitlearn that are :

# 1. train_test_split() - Randomly splits data into train/test. Works for mostly every regression problem.

# 2. StratifiedShuffleSplit()  - Splits while keeping the same category proportions (important for imbalanced data). 
                                                                                   
# StratifiedShuffleSplit is mainly for classification problems.
# In regression, we rarely use it ‚Äî except in special datasets like California housing,
# where one variable‚Äôs distribution is too uneven and must be represented evenly in both sets.
```

```python
# What is the purpose of putting algorithms on the California housing data?

# The California Housing dataset is a real-world regression problem.
# It contains features (inputs) like:

# | Feature            | Meaning                      |
# | ------------------ | ---------------------------- |
# | longitude          | How far west                 |
# | latitude           | How far north                |
# | housing_median_age | Age of houses in the area    |
# | total_rooms        | Total rooms in the block     |
# | population         | Number of people             |
# | median_income      | Median income of households  |
# | ocean_proximity    | How close it is to the ocean |

# and the label (target): median_house_value ‚Äî average price of houses in that area.
```

```python
# So what‚Äôs the purpose of applying algorithms?
# We apply Machine Learning algorithms (like Linear Regression, Decision Trees, Random Forests, etc.)
# to predict house prices based on the above features.

# This is useful for:
# üè¶ Banks or real estate firms ‚Üí estimating property values
# üèòÔ∏è Governments ‚Üí urban planning and tax assessment
# üí° Data learning ‚Üí practicing regression, feature engineering, model evaluation, etc.
# So the purpose = Build a model that can accurately predict house value for a new area based on known data.
```

```python
# Why is it important to split the data as per median_income?

# ‚ö†Ô∏è The Problem:
# If we just randomly split the dataset into train/test sets, we might end up with:

# More high-income neighborhoods in the training set,
# Fewer low-income neighborhoods in the test set.
# Then your model will learn mostly from rich areas, and fail to predict accurately for poor or middle-income areas.
# That‚Äôs called sampling bias ‚Äî and it makes your model less reliable.
```

```python
# The Solution: Stratified Split using median_income

# To fix this, we:

# Convert median_income (a continuous value) into income categories (bins) ‚Äî e.g. 1, 2, 3, 4, 5

# Use Stratified Sampling to ensure both train and test sets have the same proportion of each income category.

# This ensures:
# The training set learns from all income levels
# The test set fairly represents the same income distribution
# Your model‚Äôs performance metrics (like RMSE) are trustworthy
```
