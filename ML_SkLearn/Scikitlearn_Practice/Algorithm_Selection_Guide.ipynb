```python
# Decide What You Want to Predict - 

# 1. To Predict A category/class /label 
# like : ‚Äúspam‚Äù vs ‚Äúnot spam‚Äù, ‚Äúdisease‚Äù or ‚Äúno disease‚Äù, Iris species

# Problem type - Classification
# Algorithm Family - RandomForestClassifier, LogisticRegression, SVM, DecisionTreeClassifier, KNN

#------------------------------------------------------------------------------------------------------

# 2. To predict A continuous number 
# like : price, temperature, score, salary
# Problem Type - Regression
# Algorithm Family - LinearRegression, RandomForestRegressor, XGBoostRegressor

#------------------------------------------------------------------------------------------------------

# 3. To Predict Groups or clusters (no labels) 
# like : customer segmentation, topic grouping
# Problem Type - Clustering
# Algorithm Family - KMeans, DBSCAN, AgglomerativeClustering

#------------------------------------------------------------------------------------------------------

# 4. To predict Patterns / feature reduction
# Like : image compression, feature extraction
# Problem Type - Dimensionality Reduction
# Algorithm Family - PCA, t-SNE, LDA

#------------------------------------------------------------------------------------------------------

# 5. To predict True/False anomalies
# Like : fraud detection, outlier detection
# Problem Type - Anomaly Detection
# Algorithm Family - IsolationForest, OneClassSVM

#------------------------------------------------------------------------------------------------------
# 6 Predictions over time 
# Like : stock prices, weather
# Problem Type - Time Series Forecasting
# Algorithm Family - ARIMA, Prophet, LSTM

```

```python

#--------------------------------------------------------------------------------------------------------
# | Real-World Task             | ML Type                  | Typical Algorithm                          |
# | --------------------------- | ------------------------ | ------------------------------------------ |
# | Predict if an email is spam | Classification           | LogisticRegression, RandomForestClassifier |
# | Predict house price         | Regression               | LinearRegression, RandomForestRegressor    |
# | Group similar customers     | Clustering               | KMeans                                     |
# | Reduce 50 features to 2     | Dimensionality Reduction | PCA                                        |
# | Detect fraud or outliers    | Anomaly Detection        | IsolationForest                            |
# | Predict next month‚Äôs sales  | Time Series              | ARIMA, Prophet                             |
#--------------------------------------------------------------------------------------------------------
```

```python

#                                        RULE OF THUMB !
# ---------------------------------------------------------------------------------------------------
# | Problem Type                 | Label Type        | Example             | Use This Model         |
# | ---------------------------- | ----------------- | ------------------- | ---------------------- |
# | **Classification**           | Category / class  | Spam vs Ham         | RandomForestClassifier |
# | **Regression**               | Continuous number | House price         | LinearRegression       |
# | **Clustering**               | No labels         | Customer grouping   | KMeans                 |
# | **Dimensionality Reduction** | No labels         | Feature compression | PCA                    |
# | **Time Series**              | Time-dependent    | Stock forecast      | ARIMA                  |
# ---------------------------------------------------------------------------------------------------
```
