```python
# RMSE & MAE are for regression problems :
# They measure how far your predictions are from actual numeric values — like predicting:

# RMSE (Root Mean Squared Error) - Take the difference, square it (so big errors become even bigger), 
# find the average, and then take the square root.

# MAE (Mean Absolute Error) - Take the difference between predicted and actual values
# make them all positive (ignore minus sign) → find the average.

# Both are ways to measure how wrong your predictions are 
# — in other words, how far your model’s predictions are from the real values.
```

```python
# -----------------------------------------------------------------------------------------
# | When to use                           | Metric   | Why                             |
# | --------------------------------------| -------- | ------------------------------- |
# | You care about overall average error  |   MAE    | Simple and easy to understand   |
# | You care about avoiding big mistakes  |   RMSE   | Penalizes large errors strongly |
# -----------------------------------------------------------------------------------------
```

```python
# MAE – Mean Absolute Error

# Formula idea:
# Take the difference between predicted and actual values → make them all positive (ignore minus sign) → find the average.

# In simple words:

# “On average, how much is my prediction off by?”

# Example:
# If actual prices were [100, 200, 300]
# and predicted were [110, 190, 310]

# Errors = [10, 10, 10] →
# MAE = (10 + 10 + 10) / 3 = 10

# So your model is off by 10 units on average.
```

```python
# RMSE – Root Mean Squared Error

# Formula idea:
# Take the difference, square it (so big errors become even bigger), find the average, and then take the square root.

# In simple words: Take the difference, square it (so big errors become even bigger), find the average,
# and then take the square root.

# “It’s like MAE, but it punishes big errors more.”

# Example (same data):
# Errors = [10, 10, 10] →
# RMSE = √((10² + 10² + 10²)/3) = √(100) = 10

# But if one prediction was really bad (say, an error of 50):
# Errors = [10, 10, 50]
# RMSE = √((100 + 100 + 2500)/3) ≈ 30
# MAE = (10 + 10 + 50)/3 = 23.3

# RMSE became larger because it gives more weight to big mistakes.
```

```python
# Summary Table

# Metric            	Full Form                            Meaning           	Punishes Big Errors More      	Easier to Explain
# MAE	                Mean Absolute Error	        Average size of mistakes	             No	                      Yes

# RMSE	            Root Mean Squared Error	    Average size of mistakes 
#                                                (but large ones count more)               Yes	                  Less
```
