```python
# Using different ML Models - 

# Machine Learning (ML) is a branch of Artificial Intelligence that enables computers
# to learn from data instead of being explicitly programmed with rules.

# Why Machine Learning?
# Traditional AI programs (like Deep Blue) were good at one specific task
# (e.g., chess). ML allows us to write general-purpose programs that learn to solve many
# tasks — like playing Go, recognizing images, or filtering spam — just by learning from examples.

# AlphaGo, for instance, learned to play Go by playing against itself millions of times, improving
# its strategy without human intervention and becoming the world champion.
```

```python
# What Is a Classifier?

# A classifier is a type of ML model. It:

# Takes input data (like weight and texture of a fruit),
# Learns from examples (training data),
# And predicts labels (like "apple" or "orange").
# This process is called supervised learning.
```

```python
# Features and Labels
# In ML, we work with:

# Features: Characteristics of the data (e.g., weight, color).

# Labels: The categories we want to predict (e.g., "apple", "orange").

# If classifier is a box of rules, you can think of learning algorithm as a procedure that
# created them. It finds patterns in the training data to create a model that can predict labels for new data.
```
