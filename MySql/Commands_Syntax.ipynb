```python
# Most Common SQL Commands - Syntax and Examples - 
```

```python
# - CREATE - Creates a new table or database.

# Syntax:
# CREATE TABLE table_name (column1 datatype, column2 datatype);

# Example:
# CREATE TABLE students (id INT, name VARCHAR(50), age INT);
```

```python
# - INSERT - Adds new data into a table.

# Syntax:
# INSERT INTO table_name (column1, column2) VALUES (value1, value2);

# Example:
# INSERT INTO students (id, name, age) VALUES (1, 'Riya', 20);
```

```python
# - SELECT - Fetches data from a table.

# Syntax:
# SELECT column1, column2 FROM table_name;

# Example:
# SELECT * FROM students;
```

```python
# - DELETE - Deletes data from a table.

# Syntax:
# DELETE FROM table_name WHERE condition;

# Example:
# DELETE FROM students WHERE id = 1;
```

```python
# - UPDATE - Updates existing data.
    
# Syntax:
# UPDATE table_name SET column1 = value1 WHERE condition;

# Example:
# UPDATE students SET age = 21 WHERE name = 'Riya';
```

```python
# - ALTER - Changes table structure.
    
# Syntax:
# ALTER TABLE table_name ADD column_name datatype;

# Example:
# ALTER TABLE students ADD COLUMN grade VARCHAR(10);
```

```python
# - DROP - Deletes a table or database.

# Syntax:
# DROP TABLE table_name;

# Example:
# DROP TABLE students;
```

```python
# - WHERE - Filters data by condition.
    
# Syntax:
# SELECT * FROM table_name WHERE condition;

# Example:
# SELECT * FROM students WHERE age > 18;
```

```python
# - ORDER BY - Sorts results.

# Syntax:
# SELECT * FROM table_name ORDER BY column_name ASC|DESC;

# Example:
# SELECT * FROM students ORDER BY age DESC;
```

```python
# - GROUP BY - Groups rows for aggregate functions.

# Syntax:
# SELECT column, COUNT(*) FROM table GROUP BY column;

# Example:
# SELECT grade, COUNT(*) FROM students GROUP BY grade;
```

```python
# - HAVING - Filters grouped data.
    
# Syntax:
# SELECT column, COUNT(*) FROM table GROUP BY column HAVING condition;

# Example:
# SELECT grade, COUNT(*) FROM students GROUP BY grade HAVING COUNT(*) > 2;
```

```python
# - JOIN - Combines data from multiple tables.

# Syntax:
# SELECT t1.col, t2.col FROM t1 JOIN t2 ON t1.id = t2.id;

# Example:
# SELECT students.name, marks.score FROM students JOIN marks ON students.id = marks.student_id;
```

```python
# - DISTINCT - Removes duplicate values.
    
# Syntax:
# SELECT DISTINCT column_name FROM table_name;

# Example:
# SELECT DISTINCT grade FROM students;
```

```python
# - LIMIT - Limits number of rows shown.

# Syntax:
# SELECT * FROM table_name LIMIT number;

# Example:
# SELECT * FROM students LIMIT 5;
```

```python
# - BETWEEN / IN / LIKE - Used in WHERE for filters.

# Syntax:
# BETWEEN: SELECT * FROM students WHERE age BETWEEN 18 AND 22;

# IN: SELECT * FROM students WHERE grade IN ('A', 'B');

# LIKE: SELECT * FROM students WHERE name LIKE 'R%';
```

```python
# - TRUNCATE - Deletes all rows quickly.
    
# Syntax:
# TRUNCATE TABLE table_name;

# Example:
# TRUNCATE TABLE students;
```

```python
# - CREATE DATABASE - Creates a database.

# Syntax:
# CREATE DATABASE db_name;

# Example:
# CREATE DATABASE school;

#--------------------------------
# - USE

# Selects a database to work on.
# Syntax:
# USE db_name;
# Example:
# USE school;
```
