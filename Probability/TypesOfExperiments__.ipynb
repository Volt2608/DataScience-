```python
# What is an experiment in probability?

# An experiment is any action or process that can have two or more possible outcomes, 
# and we don‚Äôt know which will happen in advance.

# Example: Rolling a die ‚Üí you know it‚Äôll show 1‚Äì6,

# but you don‚Äôt know which number will appear.
```

```python
# üß© Types of Experiments in Probability - 

# There are mainly 3 types of probability experiments:

# 1 Deterministic
# 2Ô∏è Random
# 3Ô∏è Compound
```

```python
# 1. Deterministic Experiment - 
# The outcome is certain and predictable ‚Äî you can always tell what will happen. There‚Äôs no chance or randomness.

# Example:
# If you add 2 + 2 ‚Üí result is always 4.
# Boiling water at 100¬∞C ‚Üí always boils (under normal conditions).

# So the result is fixed and known in advance.
```

```python
# 2. Random Experiment - 
# The outcome cannot be predicted in advance, even if you know all possible results.
# You can only describe possible outcomes, not which one will occur.

# Example:
# Tossing a coin ‚Üí outcome = Heads or Tails ü™ô
# Rolling a die ‚Üí outcome = 1, 2, 3, 4, 5, or 6 üé≤
# Drawing a card from a deck üÉè
# You can‚Äôt know what‚Äôll happen until you actually perform it.

# So the result is uncertain, but from a known set of outcomes.
```

```python
# 3. Compound Experiment - 

# When you perform two or more random experiments together, it becomes a compound experiment.
# Each trial affects the total number of possible outcomes.

# Examples:
# Rolling two dice üé≤üé≤ ‚Üí 36 possible outcomes
# Tossing two coins ü™ôü™ô ‚Üí 4 possible outcomes (HH, HT, TH, TT)
# Drawing two cards from a deck (without replacement)

# The outcomes come from combining smaller random experiments.
```

```python
# Two types of outcomes in experiments - 
 
# Type	                Meaning	Example

# Single event	        Only one result	Rolling a 3
# Compound event	        Combination of results	Rolling an even number (2, 4, 6) or getting at least one head in two tosses
```

```python
# Summary chart - 
    
# Type of Experiment	      Predictable?	       Examples
# Deterministic	          ‚úÖ Yes	           2 + 2 = 4, sunrise in east
# Random	                  ‚ùå No	               Toss a coin, roll a die
# Compound	              ‚ùå No                 (but multiple randoms together)	Roll 2 dice, toss 2 coins
```

```python
# Quick tip to remember:

# If you know what will happen ‚Üí Deterministic
# If you don‚Äôt know, but it‚Äôs random ‚Üí Random
# If it‚Äôs more than one random event ‚Üí Compound
```
